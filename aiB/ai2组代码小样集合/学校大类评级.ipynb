```python
import pandas as pd
import numpy as np
import sqlite3 as db
import csv

csv_file = "res.csv"
csv_data = pd.read_csv(csv_file, low_memory = False)#防止弹出警告
data = pd.DataFrame(csv_data)
print(data)
```

```python
collegelist=list(set(data['collegeID_id']))
firstlevellist=list(set(data['firstlevelIDs']))
provincelist=list(set(data['provinceID_id']))
```

```python
newdata={}
for collegeid in collegelist:
    newdata[collegeid]={}
```

```python
for i in range(len(data)):
    collegeid=data.iloc[i]['collegeID_id']
    firstlevelid=data.iloc[i]['firstlevelIDs']
    score=data.iloc[i]['minScore']
    if firstlevelid!='None':
        if newdata[collegeid].get(firstlevelid) is None:
            newdata[collegeid][firstlevelid]={0:score}
        else:
            temp=newdata[collegeid][firstlevelid]
            temp[len(temp)]=score
```

```python
for collegeid in newdata.keys():
    school=newdata[collegeid]
    for firstlevelid in school.keys():
        firstlevel=school[firstlevelid]
        summ=0
        for score in firstlevel.values():
            summ+=score
        rankscore=summ/len(firstlevel)
        firstlevel['rankscore']=rankscore
```

```python
for collegeid in newdata.keys():
    school=newdata[collegeid]
    for firstlevelid in school.keys():
        firstlevel=school[firstlevelid]
        print(collegeid,firstlevelid,firstlevel['rankscore'])
```

```python
#按大类而不是学校编排数据
FirstLevelData={}
for collegeid in newdata.keys():
    school=newdata[collegeid]
    for firstlevelid in school.keys():
        if FirstLevelData.get(firstlevelid)==None:
            FirstLevelData[firstlevelid]={}
        if FirstLevelData[firstlevelid].get(collegeid)!=None:
            print('!!!')
        FirstLevelData[firstlevelid][collegeid]=school[firstlevelid]['rankscore']
            
```

```python
#排名
FirstLevelRank={}
for firstlevelid in FirstLevelData.keys():
    dic=FirstLevelData[firstlevelid]
    ranklist=sorted(dic.items(),key=lambda x:x[1],reverse=True)
    rank=[x[0] for x in ranklist]
    FirstLevelRank[firstlevelid]=rank
```

```python
#评估函数
def GetThreshold(length):
    a=int(0.05*length)
    b=int(0.2*length)
    c=int(0.5*length)
    d=length-1
    return a,b,c,d
        
```

```python
FirstLevelAss={}
for flid in FirstLevelRank.keys():
    FirstLevelAss[flid]={}
    rank=FirstLevelRank[flid]
    if len(rank)<=4:
        temp=['A','B','C','D']
        for i in range(len(rank)):
            FirstLevelAss[flid][rank[i]]=temp[i]
    else:
        a,b,c,d=GetThreshold(len(rank))
        for i in range(0,a+1):
            FirstLevelAss[flid][rank[i]]='A'
        for i in range(a+1,b+1):
            FirstLevelAss[flid][rank[i]]='B'
        for i in range(b+1,c+1):
            FirstLevelAss[flid][rank[i]]='C'
        for i in range(c+1,d+1):
            FirstLevelAss[flid][rank[i]]='D'
```

```python
FinalData=[]
for flid in FirstLevelAss.keys():
    for shid in FirstLevelAss[flid].keys():
        FinalData.append((flid,shid,FirstLevelAss[flid][shid]))
```

```python
title=['firstlevelIDs','collegeID_id','rank']
OutputData=pd.DataFrame(data=FinalData,columns=title)
```

```python
OutputData.to_csv("FirstLevelRank.csv", encoding="utf-8-sig")
```
