```python
import pandas as pd
reflect=pd.read_csv('result.csv')
result=pd.read_csv('分数线.csv')
#reflect.replace('图书馆,情报与档案管理','图书情报与档案管理',inplace=True)
reflect
```

```python
reflect=reflect.iloc[:,:2].to_dict(orient='records')
reflect
```

```python
reflect_dict = {}
for item in reflect:
    reflect_dict[item['major (origin)']]=item['major2 (disambiguated)']
reflect_dict
```

```python
result
```

```python
result.drop(result.iloc[:,7:],axis=1,inplace=True)
result
```

```python
first_major=pd.read_csv('一级学科(final).csv')
first_major
```

```python
major_to_ID = dict(zip(first_major.iloc[:,1],first_major.iloc[:,0]))
major_to_ID
```

```python
import re
after_disambiguate=[]
after_ID=[]
for major in result.iloc[:,4]:
    
    if major in reflect_dict.keys():
        
        temp=reflect_dict[major]
        after_disambiguate.append(temp)
        
        temp_split = re.split('[;，]',temp)
        ID_str=""
        for item in temp_split:
            ID_str =ID_str + str(major_to_ID[item]) + ';'
        after_ID.append(ID_str)
    else:
        after_disambiguate.append(major)
        after_ID.append('None')
result['after disambiguate']=after_disambiguate
result['ID']=after_ID
result.to_csv('result_after.csv')
```
